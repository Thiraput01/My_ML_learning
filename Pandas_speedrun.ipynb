# Add / Remove columns using custo func

```python
bios.head()
```

```python
# use .apply instead of using numpy
bios['height_category'] = bios['height_cm'].apply(lambda x: 'short' if x < 170 else 'tall')
bios.head()
```

```python
# apply custom function

def categorize_athlete(row):
    if row['height_cm'] < 170 and row['weight_kg'] < 70:
      return 'Lightweight'
    elif row['height_cm'] < 185 and row['weight_kg'] <= 85:
      return 'Middleweight'
    else:
      return 'Heavyweight'

bios['athlete_category'] = bios.apply(categorize_athlete, axis=1)
bios.head()
```

# Handle NULL value

```python
coffee.loc[ [0,1], 'Units Sold'] = np.nan
coffee.head()
```

```python
# count number of nans
coffee.isna().sum()
```

```python
# method 1: fill it with some num
coffee = coffee.fillna(1000)
coffee.head()
```

```python
# reccomanded using mean
coffee.loc[ [0,1], 'Units Sold'] = np.nan
coffee = coffee.fillna(coffee['Units Sold'].mean())
coffee.head()
```

```python
# or maybe interpolate (calculate the trends of data)
coffee.loc[ [0,1], 'Units Sold'] = 15
coffee.loc[ [2,3], 'Units Sold'] = np.nan
coffee = coffee.interpolate()
coffee.head()
```

```python
# method 2: drop whole row
coffee.loc[ [2,3], 'Units Sold'] = np.nan
coffee.dropna(subset='Units Sold', inplace=False) # use inplace=True for real drops
coffee.head()
```

```python
# method 3: filter data
coffee[coffee['Units Sold'].notna()]
```

```python
bios['born_date'] = pd.to_datetime(bios['born_date'])
bios['month_born'] = bios['born_date'].dt.month
bios['year_born'] = bios['born_date'].dt.year
bios.groupby( [bios['year_born'], bios['month_born']] )['name'].count().reset_index().sort_values('name', ascending=False)
```

# Advanced funcionality

```python
coffee.head()
```

```python
# shift
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
coffee.head()
```

```python
coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
coffee.head()
```

```python
# rank
bios['height_rank'] = bios['height_cm'].rank(ascending=False) # compare height on everyone
bios.sort_values(['height_rank']).head()
```

```python
# cumsum
coffee['cumulative_revenue'] = coffee['Units Sold'].cumsum()
coffee.head()
```

```python
#  rolling func
coffee['rolling_sum'] = coffee['Units Sold'].rolling(3).sum() # sum on past 3 days
coffee.head()
```
